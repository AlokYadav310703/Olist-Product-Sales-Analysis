```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## EDA

```python
cust=pd.read_csv('../data/raw/olist_customers_dataset.csv')
loc=pd.read_csv('../data/raw/olist_geolocation_dataset.csv')
orders=pd.read_csv('../data/raw/olist_orders_dataset.csv')
items=pd.read_csv('../data/raw/olist_order_items_dataset.csv')
payments=pd.read_csv('../data/raw/olist_order_payments_dataset.csv')
reviews=pd.read_csv('../data/raw/olist_order_reviews_dataset.csv')
products=pd.read_csv('../data/raw/olist_products_dataset.csv')
sellers=pd.read_csv('../data/raw/olist_sellers_dataset.csv')
category=pd.read_csv('../data/raw/product_category_name_translation.csv')
```

```python
# shape of the all csv's 
print("shape of customer dataset",cust.shape)
print("\nshape of geolocation dataset",loc.shape) 
print("\nshape of orders dataset",orders.shape) 
print("\nshape of items dataset",items.shape)
print("\nshape of payments dataset",payments.shape) 
print("\nshape of reviews dataset",reviews.shape) 
print("\nshape of products dataset",products.shape)
print("\nshape of sellers dataset",sellers.shape)
print("\nshape of category dataset",category.shape)
```

```python
# Missing values in each datsetBlockingIOError
print("\nMissing Values in customer dataset\n",cust.isnull().sum())
print("\nMissing Values in geolocation dataset\n",loc.isnull().sum()) 
print("\nMissing Values in orders dataset\n",orders.isnull().sum()) 
print("\nMissing Values in items dataset\n",items.isnull().sum())
print("\nMissing Values in payments dataset\n",payments.isnull().sum())
print("\nMissing Values in reviews dataset\n",reviews.isnull().sum()) 
print("\nMissing Values in products dataset\n",products.isnull().sum()) 
print("\nMissing Values in sellers dataset\n",sellers.isnull().sum())
print("\nMissing Values in category dataset\n",category.isnull().sum())
```

### Remove Missing Values

```python
# remove the review_comment_title and review_comment_message from reviews dataset as we don't going to use them 
reviews=reviews.drop(columns=['review_comment_title','review_comment_message','review_answer_timestamp'])
reviews.isnull().sum()
```

```python
# filling missing values in orders dataset
# Convert purchase timestamp once
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Columns you want to fill
date_columns = [
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date'
]

for col in date_columns:
    
    # Convert column to datetime
    orders[col] = pd.to_datetime(orders[col])
    
    # Compute mean delay
    mean_delay = (orders[col] - orders['order_purchase_timestamp']).mean()
    
    # Fill missing values
    orders.loc[orders[col].isnull(), col] = (
        orders['order_purchase_timestamp'] + mean_delay
    )

orders.isnull().sum()
```

```python
orders.head()
```

```python
# drop missing values from products datset 
products=products.dropna()
products.isnull().sum() 
```

```python
# distribution of review score 
sns.countplot(x='review_score', data=reviews)
plt.title("Distribution of Review Scores")
plt.show()
```

```python
# delivery time delays 
orders['delivery_time'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days
sns.histplot(orders['delivery_time'], bins=50, kde=True)
plt.title("Delivery Time Distribution")
plt.show()
```

```python
# delivery time delays 
orders['delivery_time'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days
orders['delivery_time']=np.log1p(orders['delivery_time']) #taking log transformation to remove skewness
sns.histplot(orders['delivery_time'], bins=50, kde=True)
plt.title("Delivery Time Distribution")
plt.show()
```

## Save all the Processed CSV's

```python
cust.to_csv('../data/processed/olist_customers_dataset.csv', index=False)
loc.to_csv('../data/processed/olist_geolocation_dataset.csv', index=False)
orders.to_csv('../data/processed/olist_orders_dataset.csv', index=False)
items.to_csv('../data/processed/olist_order_items_dataset.csv', index=False)
payments.to_csv('../data/processed/olist_order_payments_dataset.csv', index=False)
reviews.to_csv('../data/processed/olist_order_reviews_dataset.csv', index=False)
products.to_csv('../data/processed/olist_products_dataset.csv', index=False)
sellers.to_csv('../data/processed/olist_sellers_dataset.csv', index=False)
category.to_csv('../data/processed/product_category_name_translation.csv', index=False)
```

## Merge all processed CSV's into single CSV

```python
merged=cust[['customer_id','customer_unique_id','customer_state','customer_city']].merge(
    orders, 
    on='customer_id', 
    how='inner'
)

merged=merged.merge(
    items.drop('order_item_id', axis=1),
    on='order_id', 
    how='inner'
)

merged=merged.merge(
    payments,
    on='order_id', 
    how='inner'
)

merged=merged.merge(
    reviews,
    on='order_id', 
    how='inner'
)

merged=merged.merge(
    products[['product_id','product_category_name','product_weight_g']],
    on='product_id', 
    how='inner'
)

merged=merged.merge(
    sellers.drop('seller_zip_code_prefix', axis=1),
    on='seller_id', 
    how='inner'
)
```
