```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Merged_dataset.csv')
```

### Sellers average rating

```python
avg_seller_rating=(
    df
    .groupby('seller_id')['review_score'].mean().reset_index()
)

avg_seller_rating.rename(columns={'review_score': 'mean_seller_rating'}, inplace=True)
avg_seller_rating
```

```python
bad_performing_sellers=avg_seller_rating[avg_seller_rating['mean_seller_rating']>2.5]
bad_performing_sellers.sort_values(by='mean_seller_rating').head(5)
```

### Number of orders processed by each seller

```python
total_orders_processed=(
    df
    .groupby('seller_id')
    .agg(
        total_orders=('order_id','count')
    )
)
total_orders_processed=total_orders_processed.merge(
    avg_seller_rating, 
    on='seller_id', 
    how='left'
)
```

```python
total_orders_processed.sort_values(by='total_orders', ascending=False).head()
```

```python
df['order_purchase_timestamp']=pd.to_datetime(df['order_purchase_timestamp'])
df['year']=df['order_purchase_timestamp'].dt.year
df['months']=df['order_purchase_timestamp'].dt.month

all_months = range(1, 13)
df['months'] = pd.Categorical(
    df['months'],
    categories=all_months,
    ordered=True
)

df=df.merge(
    total_orders_processed, 
    on='seller_id', 
    how='inner'
)
```

## Function to display Seller stats

```python
def seller_stat(seller_id):
    seller_data=df[df['seller_id']==seller_id]

    category_count = (
        seller_data
        .groupby(['seller_id', 'product_category_name'])
        .agg(
            total_units=('product_category_name', 'count')
        )
        .reset_index()
    )

    time_frame_sales = (
            seller_data
            .groupby(['year',
                      'product_category_name',
                      'months'])
            .size()
            .unstack(fill_value=0) # without this purchase month would be a column 
            .reset_index()
    )
    time_frame_sales['total_sales'] = time_frame_sales.loc[:, 1:12].sum(axis=1) # total sales of particular category in a year
    time_frame_sales.columns.name = None # remove the name of the column index created after unstack(). (month)

    total_yearly_sales=seller_data['year'].value_counts().reset_index().sort_values(by='year') # total sales in a particular year

    seller_rating=seller_data['review_score'].unique()

    # print(category_count)
    return time_frame_sales, total_yearly_sales, seller_rating   
```

```python
# input a seller ID to check Seller Details
time_frame_sales, total_yearly_sales, seller_rating=seller_stat('6560211a19b47992c3666cc44a7e94c0')

print(f"-----------Total monthly sales by category------------- \n\n{time_frame_sales}")
print(f"\n---------Total yearly sales-------------------------- \n{total_yearly_sales}")
print(f"\n---------Seller Rating------------------------------- \n{seller_rating}")

pivot_df = time_frame_sales.pivot(
    index='year',
    columns='product_category_name',
    values='total_sales'
)

pivot_df.plot(kind='bar')

plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Year vs Category Sales")
plt.show()
```

### late delivery ratio by each seller

```python
df["late_delivery"] = (
    pd.to_datetime(df["order_delivered_customer_date"])
    - pd.to_datetime(df["order_estimated_delivery_date"])
).dt.days
```

```python
df['late_delivery']=df['late_delivery']>0
```

```python
late_del_count=(
    df
    .groupby('seller_id')
    .agg(
        late_del=('late_delivery',lambda x: (x==True).sum()),
        on_time=('late_delivery',lambda x: (x==False).sum()), 
        total_del=('late_delivery','count'),
    )
)
```

```python
late_del_count.reset_index()
late_del_count['ratio']=late_del_count['late_del']/late_del_count['total_del']
late_del_count['ratio'].max()
```
